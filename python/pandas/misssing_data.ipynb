# Missing data 
- Find missing data and fill it with a specific value or drop the rows/columns with missing data.

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', np.nan, 'Charlie', 'David'],
    'Age': [25, 30, np.nan, 40],
    'City': ['New York', 'Los Angeles', np.nan, np.nan],
    'Designation': [np.nan, 'Manager', 'Analyst', 'Director']
}
```

```python
df = pd.DataFrame(data) 
print(df)
```

```python
df.isna() # Check for missing values
```

```python
df.isna().sum() # Count non-missing values in each column
```

```python
df.isnull() # Check for missing values
df.isna().any() # Check if any missing values exist in the DataFrame
```

## Removing missing data

```python
df.dropna() # Drop rows with missing values
```

```python
df.dropna(thresh=3) # Drop rows with less than 3 non-missing values
```

## Filling missing data

```python
df.fillna('Unknown') # Fill missing values with 'Unknown'
```

```python
values = {'Name': 'Unknown', 'Age': df['Age'].mean(), 'City': 'Unknown', 'Designation': 'Unknown'}
df.fillna(value=values) # Fill missing values with specified values
```

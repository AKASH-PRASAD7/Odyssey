# Pandas Series
- A Pandas Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It is similar to a column in a spreadsheet or a SQL table. Each element in a Series has an associated label, which can be used to access the data.

```python
import numpy as np
import pandas as pd
```

## Series Creation

```python
labels = ['a','b', 'c', 'd']
list = [10, 20, 30, 40]
arr = np.array(list)
d = {'a':10, 'b':20, 'c':30, 'd':40}

series = pd.Series(list)

print("Series \n", series)
```

```python
print("Series with labels \n \n",pd.Series(list, index=labels))
```

```python
print("Series with array \n",pd.Series(arr))
```

```python
print("Series with Dictonary \n",pd.Series(d))
```

## Operations on DataFrames

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
```

```python
df.shape # dimensions of the dataframe(matrix)
```

```python
df.describe() # statistical summary of the dataframe
```

```python
df.columns
```

```python
df.info # summary of the dataframe, including data types and non-null counts
```

```python
df['A'] *2 # element-wise multiplication of column 'A' by 2
```

```python
df['D'] = df["B"].apply(lambda x : x**2)
df
```

# DataFrame
- A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table. Each column in a DataFrame can be thought of as a Series, and the entire DataFrame can be thought of as a collection of Series that share the same index.

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [70000, 80000, 90000, 100000]
}
```

```python
data_frame = pd.DataFrame(data)
print(data_frame)
```

```python
data_list = [
    ['Alice', 25, 'New York', 70000],
    ['Bob', 30, 'Los Angeles', 80000],
    ['Charlie', 35, 'Chicago', 90000],
    ['David', 40, 'Houston', 100000]
]
columns=['Name', 'Age', 'City', 'Salary']
data_frame = pd.DataFrame(data_list)
data_frame2 = pd.DataFrame(data_list, columns=columns)
print(data_frame)
print("\nDataFrame with columns \n")
print(data_frame2)
```

## Selecting/ Indexing Colums

```python
data_frame2["Name"]
```

```python
data_frame2[['Name','City']]
```

## Adding a Colums

```python
data_frame2['Designation']  = ['Reciptionist','Teacher','Athlete','Plumber']
print(data_frame2)
```

## Removing Columns

```python
data_frame2.drop('Designation', axis=1) #  Won't change the original data frame just a view
# Axis = 1 for columns and axis = 0 for rows
```

```python
data_frame2.drop(['Salary','Designation'], axis=1, inplace=True) # Wiil modify the original data frame
print(data_frame2) 
```

## Selecting Rows

```python
data_frame2.loc[0] # Selecting the first row
```

```python
data_frame2.loc[[0,2]] # Selecting the first and third row
```

```python
data_frame2.iloc[0] # Selecting the first row using index
```

## Selecting portions of the DataFrame

```python
data_frame2.loc[[0,1],['Name','City']] # Selecting specific rows and columns
```

```python
data_frame2.loc[[2,3],['City','Designation']] # Selecting specific rows and columns
```

## Conditional Selection

```python
# Select where age is greater than 30
data_frame2[data_frame2['Age'] > 30]
```

```python
# Select where age is greater than 30 and city is Chicago
data_frame2[(data_frame2['Age'] > 30) & (data_frame2['City'] == 'Chicago')]
```

# Joining dataframes


```python
import numpy as np
import pandas as pd
```

```python

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
})

print(users)

user_cities = pd.DataFrame({
    'id': [1, 2, 3, 6,7],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
})
print(user_cities)

```

```python
pd.merge(users, user_cities, on='id', how='inner')
```

```python
pd.merge(users, user_cities, on='id', how='outer')
```

```python
pd.merge(users, user_cities, on='id', how='left')
```

```python
pd.merge(users, user_cities, on='id', how='right')
```

## Concatenation
Concatenation is the process of combining two or more dataframes along a particular axis (either rows or columns). In pandas, you can use the `pd.concat()` function to concatenate dataframes.

```python
fruits = pd.DataFrame({
    'color': ['Red', 'Yellow', 'Orange'],
    'fruit': ['Apple', 'Banana', 'Mango']
})
fruits1 = pd.DataFrame({
    'color': ['Red', 'Yellow', 'Green'],
    'fruit': ['Apple', 'Banana', 'Kiwi']
})
```

```python
pd.concat([fruits,fruits1]) 
```

```python
pd.concat([fruits,fruits1], axis=1) 
```

## Joining

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4]) 

df1
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```

## List vs Numpy Array
- Lists are general-purpose containers that can hold different types of data, while NumPy arrays are specifically designed for numerical computations and are more efficient for large datasets.

```python
import time
list = [1,2,3,4]
arr = np.array(list)

print("Python list multiplication:", list * 2) 
print("NumPy array multiplication:", arr * 2)

start = time.time()

l1 = [x*2 for x in range(1000000)]
print("\nTime taken for list comprehension:", time.time() - start)

start = time.time()
a1 = np.arange(1000000) * 2
print("\nTime taken for NumPy array operation:", time.time() - start)
```

## Array Creation

```python
zeros = np.zeros((3, 4))
print("zeros array: \n", zeros)

ones = np.ones((2, 3))
print("one array: \n", ones)

full = np.full((2, 2), 7)
print("full array: \n", full)

random = np.random.random((2, 3))
print("random array: \n", random)

sequence = np.arange(0, 11, 2)
print("sequnce array: \n", sequence)
```

## Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1, 2], [3, 4]],  # Multidimensional arrays are called tensors
                   [[5, 6], [7, 8]]])
print("Tensor: ", tensor)
```

## Array properties

```python
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])
print("Shape ", arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("DType ", arr.dtype)
```

## Array Reshaping


```python
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3, 4))
print("\n Reshaped array ", reshaped)

flattened = reshaped.flatten()
print("\n Flattened array ", flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array ", transpose)
```
